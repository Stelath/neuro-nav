```python
import time
import math
import sdl2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

from brainflow.board_shim import BoardShim, BrainFlowInputParams, LogLevels, BoardIds
from brainflow.data_filter import DataFilter, AggOperations, FilterTypes, WindowOperations, NoiseTypes

from stable_baselines3 import PPO
from eeg_ccn_extractor import EEGCNNExtractor
from kinetic_gym import KineticEnv
```

```python
policy_kwargs = dict(
    features_extractor_class=EEGCNNExtractor,
    features_extractor_kwargs=dict(features_dim=256),
)

env = KineticEnv(render=True)
model = PPO('CnnPolicy', env, policy_kwargs=policy_kwargs, verbose=1)
```

```python
def callback(*args):
    env.render()

model.learn(total_timesteps=2048, callback=callback, tb_log_name="runs")
```

```python
model.save("ppo_kinetic.pth")
```

```python
env.close()
```
